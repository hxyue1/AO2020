```python
outcomes_df = pd.DataFrame(outcomes)
outcomes_df.columns = ['player_1', 'player_2', 'player_1_win']
outcomes_df.loc[:,'player_1'] = outcomes_df['player_1'].str.lower()
outcomes_df.loc[:,'player_2'] = outcomes_df['player_2'].str.lower()
```

```python
outcomes_df.to_csv('Data/2020_outcomes.csv')
```

```python
apriori_preds = pd.read_csv('Predictions/a-priori-predictions.csv')
apriori_preds_corrected = pd.read_csv('Predictions/a-priori-predictions-corrected.csv')
expost_preds = pd.read_csv('Predictions/ex-post-predictions.csv')
logit_rank_only_preds = pd.read_csv('Predictions/logit_rank_only.csv')
true_naive_preds = pd.read_csv('Predictions/true_naive.csv')
```

```python
results_apriori = outcomes_df.merge(apriori_preds,
                            how='left',
                            left_on = ['player_1', 'player_2'],
                            right_on = ['player_1', 'player_2'],
                            validate = 'm:1')
results_apriori.fillna(0.5, inplace = True)
results_apriori_corrected = outcomes_df.merge(apriori_preds_corrected,
                            how='left',
                            left_on = ['player_1', 'player_2'],
                            right_on = ['player_1', 'player_2'],
                            validate = 'm:1')

results_expost = outcomes_df.merge(expost_preds,
                            how='left',
                            left_on = ['player_1', 'player_2'],
                            right_on = ['player_1', 'player_2'],
                            validate = 'm:1')
results_logit_rank_only = outcomes_df.merge(logit_rank_only_preds,
                            how='left',
                            left_on = ['player_1', 'player_2'],
                            right_on = ['player_1', 'player_2'],
                            validate = 'm:1')
results_true_naive = outcomes_df.merge(true_naive_preds,
                            how='left',
                            left_on = ['player_1', 'player_2'],
                            right_on = ['player_1', 'player_2'],
                            validate = 'm:1')
```

```python
a_priori_metrics = [log_loss(results_apriori['player_1_win'], results_apriori['player_1_win_probability']), 
      roc_auc_score(results_apriori['player_1_win'], results_apriori['player_1_win_probability']),
      accuracy_score(results_apriori['player_1_win'], np.round(results_apriori['player_1_win_probability']))]
```

```python
a_priori_corrected_metrics = [log_loss(results_apriori_corrected['player_1_win'], results_apriori_corrected['player_1_win_probability']), 
      roc_auc_score(results_apriori_corrected['player_1_win'], results_apriori_corrected['player_1_win_probability']),
      accuracy_score(results_apriori_corrected['player_1_win'], np.round(results_apriori_corrected['player_1_win_probability']))]
```

```python
ex_post_metrics = [log_loss(results_expost['player_1_win'], results_expost['player_1_win_probability']),
      roc_auc_score(results_expost['player_1_win'], results_expost['player_1_win_probability']),
      accuracy_score(results_expost['player_1_win'], np.round(results_expost['player_1_win_probability']))]
```

```python
logit_rank_only_metrics = [log_loss(results_logit_rank_only['player_1_win'], results_logit_rank_only['player_1_win_probability']),
      roc_auc_score(results_logit_rank_only['player_1_win'], results_logit_rank_only['player_1_win_probability']),
      accuracy_score(results_logit_rank_only['player_1_win'], np.round(results_logit_rank_only['player_1_win_probability']))]
```

```python
true_naive_rank_only_metrics = [log_loss(results_true_naive['player_1_win'], results_true_naive['player_1_win_probability']),
      roc_auc_score(results_true_naive['player_1_win'], results_true_naive['player_1_win_probability']),
      accuracy_score(results_true_naive['player_1_win'], np.round(results_true_naive['player_1_win_probability']))]
```

```python
results = pd.DataFrame([a_priori_metrics, a_priori_corrected_metrics, ex_post_metrics, logit_rank_only_metrics, true_naive_rank_only_metrics])
results.index = ['a-priori', 'a-priori-corrected', 'ex-post', 'logit-rank-only', 'indicator-rank-only']
results.columns = ['logloss', 'AUC', 'accuracy']
```
